```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=pd.read_csv('Algerian_forest_fires_dataset_UPDATE.csv',header=1)
```

```python
data.info()
```

Data Cleaning

```python
# finding missing values

data[data.isnull().any(axis=1)]
```

```python
data.loc[:122,"Region"]=0
data.loc[122:,"Region"]=1
```

```python
data.info()
```

```python
data[['Region']]=data[['Region']].astype(int)
```

```python
data.head()
```

```python
data.isnull().sum()
```

```python
#droping all null values

data=data.dropna().reset_index(drop=True)
```

```python
data.head()
```

```python
data.isnull().sum()
```

```python
data.iloc[[122]]
```

```python
data=data.drop(122).reset_index(drop=True)
```

```python
data.iloc[[122]]
```

```python
data.columns
```

```python
#fix space in column name
data.columns=data.columns.str.strip()
```

```python
data.columns
```

Change the required columns as integer data type

```python
data[['day', 'month', 'year', 'Temperature', 'RH', 'Ws']]=data[['day', 'month', 'year', 'Temperature', 'RH', 'Ws']].astype(int)
```

```python
data.info()
```

Changing the required column as float data type

```python
data[['Rain', 'FFMC','DMC', 'DC', 'ISI', 'BUI', 'FWI',]]=data[['Rain', 'FFMC','DMC', 'DC', 'ISI', 'BUI', 'FWI',]].astype(float)
```

```python
data.info()
```

```python
data.describe()
```

Save the cleaned dataset

```python
data.to_csv('Algerian_forest_fires_Cleaned_dataset.csv',index=False)
```

Exploratory Data Analysis

```python
#drop day ,month and year columns
data_copy_1=data.drop(['day','year','month'],axis=1)
```

```python
data_copy_1.head()
```

Perform Encoding of Classes Column

```python
data_copy_1['Classes']=np.where(data_copy_1['Classes'].str.contains('not fire'),0,1)
```

```python
data_copy_1.head()
```

```python
data_copy_1['Classes'].value_counts()
```

plot density plot for all features

```python
plt.style.use("bmh")
data_copy_1.hist(bins=50,figsize=(20,16))
plt.show()
```

```python
percentage=data_copy_1['Classes'].value_counts(normalize=True)*100
```

Polting Pie Chart

```python
classlabels=['Fire','Not Fire']
plt.figure(figsize=(12,7))
plt.pie(percentage,labels=classlabels,autopct="%1.1f%%")
plt.title("Pie Chart of Classes")
plt.show()
```
